```python
d1 = pd.read_csv('data/Traffic_Crashes_-_Crashes.csv')
d2 = pd.read_csv('data/Traffic_Crashes_-_People.csv')
d3 = pd.read_csv('data/Traffic_Crashes_-_Vehicles.csv')
```

```python
# Combine d1, d2, and d3 into a single dataframe on CRASH_RECORD_ID
d = d1.merge(d2, on='CRASH_RECORD_ID').merge(d3, on='CRASH_RECORD_ID')
```

```python
# split d into categorical and numerical dataframes
d_cat = d.select_dtypes(include='object')
d_num = d.select_dtypes(exclude='object')
```

```python
d_cat.isna().sum()

# export the number of nulls for each column to a csv
d_cat.isna().sum().to_csv('data/cat_nulls.csv')
```

```python
# drop all columns with more than 24 nulls
d_num.dropna(axis=1, inplace=True)

# check
d_num.isnull().sum()
```

```python
# make a copy of d_num
d_num2 = d_num.copy()
```

```python
d_num.head()
```

```python
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split

# Define the preprocessing steps
preprocessing_steps = [
    ('scaling', StandardScaler())
]

# Create the pipeline
pipeline = Pipeline(preprocessing_steps)

# Fit the pipeline on the training data
d_num2_processed = pipeline.fit_transform(d_num2)
```
